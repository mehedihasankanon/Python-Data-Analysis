## What is DataFrame? 
---
 
- A DataFrame is the most important data structure in Pandas. 
- It is a 2-dimensional labeled table where:
    - Rows represent observations/records.
    - Columns represent features/variables.
    - Each **column** is actually a Pandas **Series**.

## Creating DataFrame 
---

- using `pd.DataFrame()` constructor. 

| Method               | Input                |   
| -------------------- | -------------------- | 
| Dictionary           | dict of lists/arrays | 
| List of dictionaries | list of dict         | 
| NumPy array          | ndarray              | 
| List of lists        | nested list          | 
| Reading files        | CSV, Excel, SQL      | 


```python
# example: 
data = {
    "Name": ["Rafi", "Kanon", "Sami", "Barson", "Maskat"], 
    "City": ["Chattogram", "Mymensingh", "Natore", "Khulna", "Bogra"], 
    "Age": [30, 22, 25, 21, 20], 
    "CGPA": [3.40, 3.92, 3.66, 3.89, 3.71] 
}   # dict of lists 

df = pd.DataFrame(data)
# df = pd.DataFrame(data, index=["A", "B", "C", "D", "E"])     # custom indexing of rows

print(df)
```

```python
# from numpy array 

arr = np.array(
    [
        ["Rafi", "Chattogram", 3.5], 
        ["Kanon", "Dhaka", 3.88], 
        ["Sami", "Rajshahi", 3.76]
    ]
) 

df1 = pd.DataFrame(arr, columns=["Name", "City", "CGPA"])    # custom column names 
print(df1)
```

## Rows, Columns, Index, and Label 
---



| Component     | Meaning                |
| ------------- | ---------------------- |
| Row           | One observation/record |
| Column        | One feature/variable   |
| Index         | Labels of rows         |
| Column labels | Names of columns       |


```python
# example: 
data = {
    "Name": ["Rafi", "Kanon", "Sami", "Barson", "Maskat"], 
    "City": ["Chattogram", "Mymensingh", "Natore", "Khulna", "Bogra"], 
    "Age": [30, 22, 25, 21, 20], 
    "CGPA": [3.40, 3.92, 3.66, 3.89, 3.71] 
}   # dict of lists 

df1 = pd.DataFrame(data) 
# print(df1) 

# access a column 
print(df1["Name"], "\n")

# access multiple columns 
print(df1[["Name", "City"]])     # pass a list of columns 
```

```python
# all columns 
print(df1.columns, "\n") 

# index range 
print(df1.index, "\n")

# all info 
print(df1.info(), "\n") 

# describe 
print(np.round(df1.describe(), 2), "\n")     # statistical summary
```

## `loc` and `iloc` and Slicing
---

| Method | Uses              | End Index |
| ------ | ----------------- |----------|
| `loc`  | Labels            | inclusive |
| `iloc` | Integer positions | exclusive |


```python
# example 
print(df, "\n")

# integer locations
print(df.iloc[0], "\n") 
print(df.iloc[-1], "\n")    # last row 
print(df.iloc[1:3], "\n")
print(df.iloc[1:3, 2:4], "\n")      # [row_slice, col_slice] 
```

```python
# labels 
df1 = df.copy()
df1.index = ['A', 'B', 'C', 'D', 'E']   # change the row idices 
print(df1, "\n")

print(df1.loc['B'], "\n")
print(df1.loc['B':'D'], "\n")
print(df1.loc['B':'D', "City":"Age"], "\n")     # (row_slice, col_slice)
```

## Attributes of DataFrame
---

| Attribute | Meaning                  |
| --------- | ------------------------ |
| `shape`   | (rows, columns)          |
| `size`    | Total elements           |
| `ndim`    | Number of dimensions     |
| `columns` | Column labels            |
| `index`   | Row labels               |
| `dtypes`  | Data type of each column |
| `values`  | NumPy array              |
| `T`       | Transpose                |


```python
# example 
print(df, "\n") 

print(df.shape)
print(df.size)
print(df.ndim, "\n")
print(df.dtypes, "\n")
print(df.columns, "\n")     # type = pandas.Index
print(df.index, "\n")       # type = pandas.RangeIndex
print(df.values, "\n")      # type = numpy.ndarray
```

## DataTypes and Type Casting 
---

| dtype      | Meaning             |
| ---------- | ------------------- |
| int64      | Integer             |
| float64    | Decimal             |
| object     | Text (older pandas) |
| string     | String dtype        |
| bool       | Boolean             |
| datetime64 | Date & Time         |


- each DataFrame column can have a different data type.
- use `astype(dtype)` function for casting. 

```python
# example 
print(df, "\n") 
print(df.dtypes, "\n") 

df["Age"] = df["Age"].astype(float)     # type casting 
print(df["Age"].dtype)
```

## Adding, Updating, and Deleting Rows/Columns
---

### Adding Rows 
--- 

- create a new DataFrame with the new rows
- **concatenate** the new DataFrame with the actual one 

```python
# example 
print(df, "\n") 

new_data = {
    "Name": ["Muktadir", "Mohtasim", "Tamim"], 
    "City": ["Dinajpur", "Rajshahi", "Chattogram"], 
    "Age": [30, 22, 25], 
    "CGPA": [3.77, 3.94, 3.89] 
}   # dict of lists 

new_df = pd.DataFrame(new_data) 

df = pd.concat([df, new_df], ignore_index = True)  
print(df)
```

```python
# example using loc[] 
print(df.columns) 
print("Length of the DataFrame:", len(df))  # number of rows 
df.loc[len(df)] = ["Ahnaf", "Dhaka", 22, 3.91] 
print(df)
```

### Delete a Row 
--- 

- use `df.drop(index= ...)` to delete rows. 
- use `inplace = True` to modify the original DataFrame directly. 

```python
# example 
print(df, "\n") 

df = df.drop(index= 9) 
# df.drop(index=[8, 9], inplace= True)   

print(df)
```
