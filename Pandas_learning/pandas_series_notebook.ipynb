## What is Pandas? 
---

- Pandas = Panel Data
- Pandas is a Python library used for working with structured data (tables, spreadsheets, CSV files, databases, etc.). 
- Pandas is like Excel inside Python. 
- Main data structures of Pandas: 
    1. **Series:** a one-dimensional labeled array. 
    1. **DataFrame:** a two-dimensional table. 

## Applications of Pandas
---

- Load data from CSV, Excel, SQL databases, JSON, APIs
- Explore data 
- Clean data 
- Filter data 
- Analyze data 
    - Sort data 
    - Group data  
- Transform data 

## NumPy vs Pandas
---

| Feature | NumPy | Pandas | 
|---------|-------|--------| 
| purpose | numerical computing | data analysis |
| data type | ndarray | Series, DataFrame | 
| labels <br> (row, column names) | no | yes |  
| handles tabular data | limited | excellent | 
| missing values support | basic | strong |  
| ML/Data analysis | foundation | daily tool | 

- NumPy = fast numerical arrays 
- Pandas = data tables built on top of NumPy

### Creating Series 
---

```python
# use Series() constructor to create Series object 
# can create Series from list, tuple, dictonaries. 
# but set are not allowed, cause sets are unordered 

data = [10, 20, 30, 40] 
series = pd.Series(data) 

print(series)
```

```python
# other data types 

fruits = ('apple', 'banana', 'orange') 
bool_data = [True, False, True, True] 

tuple_to_series = pd.Series(fruits) 
list_to_series = pd.Series(bool_data) 

print(tuple_to_series) 
print(list_to_series)
```

```python
# from dictionary 

marks = {"Rafi": 16, "Sami": 16, "Kanon": 17} 
series = pd.Series(marks) 

print(series)
```

### Custom Index/Label
---

```python
data = [100, 200, 300] 
series = pd.Series(data, index=['a', 'b', 'c']) 

print(series)
```

### Accessing values using labels
---

```python
data = [10, 20, 30] 
fruits = ('apple', 'banana', 'mango', 'orange') 
marks = {"Rafi": 16, "Sami": 16, "Kanon": 17} 

data_series = pd.Series(data, index=('a', 'b' ,'c')) 
fruits_series = pd.Series(fruits) 
marks_series = pd.Series(marks)

# access values using index keys: 
fruits_series[3] = 'lichi' 
print(data_series['b'])
print(fruits_series[3]) 
print(marks_series["Rafi"])
```

### `loc` and `iloc`
---

```python
data = [10, 20, 30] 
fruits = ('apple', 'banana', 'mango', 'orange') 
marks = {"Rafi": 16, "Sami": 16, "Kanon": 17} 

data_series = pd.Series(data, index=('a', 'b' ,'c')) 
fruits_series = pd.Series(fruits) 
marks_series = pd.Series(marks) 

# using position (iloc = integer location) 
print(marks_series.iloc[1]) 
print(fruits_series.iloc[-1])   # negative indexing 

# using label (loc = location by label) 
print(data_series.loc['b']) 

marks_series.loc["Kanon"] = 19
print(marks_series.loc["Kanon"])
```

### Series Attributes
---

| Attribute | Meaning | Returns |
|-----------|---------|---------|
| values | actual data stored | NumPy Array |
| index | labels of the data | Range Index | 
| dtype | data type of values | single result <br> (int64, float64, string, object, bool) | 
| shape | dimensions of the Series | tuple | 
| size | total number of elements | single result |
| ndim | number of dimensions | single result | 

```python
data1 = [10, 15, 20, 25, 30] 
data2 = ['apple', 'banana', 'mango', 'orange'] 
data3 = [True, False, True, True] 
data4 = [10.1, 15.67, 21.33] 
data5 = [10, 'apple', True, 12.59] 

s1 = pd.Series(data1) 
s2 = pd.Series(data2)
s3 = pd.Series(data3)
s4 = pd.Series(data4)
s5 = pd.Series(data5)
```

```python
# print data types of the values stored 
print(s1.dtype) 
print(s2.dtype) 
print(s3.dtype) 
print(s4.dtype) 
print(s5.dtype) 
```

```python
# print shape of the Series
print(s1.shape) 
print(s2.shape) 
print(s3.shape) 
print(s4.shape) 
print(s5.shape) 
```

```python
# print size of the Series 
print(s1.size)
print(s2.size)
print(s3.size)
print(s4.size)
print(s5.size)
```

```python
# print number of dimensions of the Series
print(s1.ndim)
print(s2.ndim)
print(s3.ndim)
print(s4.ndim)
print(s5.ndim)
```

```python
# print all values of a Series as NumPy array:
print(s1.values) 
print(type(s1.values), "\n") 

print(s2.values) 
print(type(s2.values), "\n")

print(s3.values) 
print(type(s3.values), "\n")

print(s4.values) 
print(type(s4.values), "\n")

print(s5.values) 
print(type(s5.values), "\n")
```

```python
# print index range (integer indexing) of a Series as Range object:
print(s1.index) 
print(type(s1.index), "\n") 

print(s2.index) 
print(type(s2.index), "\n")

print(s3.index) 
print(type(s3.index), "\n")

print(s4.index) 
print(type(s4.index), "\n")

print(s5.index) 
print(type(s5.index), "\n")
```

### Vectorized Operations
---

List of operations: 
- addition 
- subtraction 
- multiplication 
- division 
- modulus 
- power 
- logical operations 

```python
list1 = [10, 20, 30, 40] 
list2 = [1.56, 2.91, -3.75, 0.84]

s1 = pd.Series(list1) 
s2 = pd.Series(list2) 
```

```python
# scalar operations: 
print(s1 + 5)
print(s1 - 5)
print(s1 * 5)
print(s1 / 5)
print(s1 % 5)
print(s1 ** 2)

# logical operators: 
print(s1 > 25) 
print((s1>10) & (s1<35))
```

```python
# vectorized operations: 
print(s1 + s2) 
print(s1 - s2) 
print(s1 * s2) 
print(s1 / s2) 
print(s1 ** s2)  

# logical operation
print(s1 > s2)
```

```python
# rounding operations: 
print(s2.round(1)) 
print(s2.abs()) 
```

### Slicing 
---

**Syntax:** `series[start:stop]`

- `position-based` slicing (stop excluded)
- `label-based` slicing (stop included)

```python
s = pd.Series([10, 20, 30, 40, 50], index=['a', 'b', 'c', 'd', 'e'])

# position-based slicing 
print(s.iloc[1:4])       # index 4 excluded 
# print(s[1:4])     # same result 

# label-based slicing 
print(s.loc['b':'e'])   # index 'e' included 
# print(s['b':'e'])     # same result 
```

### Boolean Indexing
---

```python
s = pd.Series([10, 20, -1, -5, 0, 100, 89, 65]) 

print(s[s>=0]) 
print(s[(s>50) | (s<0)])
```

### Sorting Series 
---

**Two Ways:** 
1. sort by values 
1. sort by index

```python
marks = {
    "Rafi": 49, "Sami": 67, "Kanon": 76, 
    "Tamim": 55, "Barson": 89, "Maskat": 61
}

s = pd.Series(marks) 

print(s.sort_values(), "\n")
print(s.sort_index(), "\n")

# descending order: 
print(s.sort_values(ascending = False))
```

### Unique Values
---

```python
# prepare sample: 
data = [i for i in range(1, 21)] 
data.extend(list(i for i in range(1, 21, 2))) 
data.sort()

# create series: 
s = pd.Series(data) 

print(s.unique())   # unique values only (as NumPy array)
print("number of unique values:", s.nunique()) 
print(s.value_counts())     # count appearance of each value
```

### Missing Values 
---
**NaN** = Not a Number 

```python
# NumPy array to Pandas Series creation: 
arr = np.array([10, 20, np.nan, 40, np.nan, 60, 70, np.nan]) 
s = pd.Series(arr)
print(s)
```

```python
# detect missing values: 
print(s[s.isna()]) 
print(s[s.isnull()]) 
```

```python
# count missing values: 
print(s.isna().sum()) 
```

```python
# drop missing values: 
print(s.dropna())
```

```python
# fill missing values: 
print(s.fillna(0))
```

### Statistical Functions 
---
- sum 
- count 
- max 
- min 
- mean 
- meadian 
- standard deviation 
- variance 
- describe

```python
data = [1, 20, -5, 67, 92, 8, -11, 34, 3, 0, -16] 
s = pd.Series(data) 

print(s, "\n")
print("sum:", s.sum())
print("count:", s.count())
print("max:", s.max())
print("min:", s.min()) 
print("mean:", s.mean()) 
print("median:", s.median()) 
print("mode:\n", s.mode(), sep="") 
print("standard deviation:", s.std()) 
print("variance:", s.var()) 
```

```python
# quick statistical summary of data: 
print(s.describe())
```

### Conversion
---

**Anything to Series Conversion:**

use `Series()` constructor to convert 
- list 
- tuple 
- dictionary 
- NumPy array 

to Pandas Series object.  

**Series to others conversion:**

use the built in functions: 
- to_numpy()
- to_list() 
- to_dict() 
- to_csv() 
- to_string() 

and many more...

```python
# series to NumPy array conversion: 
s = pd.Series([10, 20, 30, 40])
arr = s.to_numpy() 

print(arr)
print(type(arr)) 
```

```python
# series to list:
s = pd.Series([10, 20, 30, 40])
mylist = s.to_list() 

print(mylist)
print(type(mylist))
```

```python
# series to dictionary:
s = pd.Series([10, 20, 30, 40])
mydict = s.to_dict() 

print(mydict)
print(type(mydict))
```
