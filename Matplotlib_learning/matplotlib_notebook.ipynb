## Title, Labels, Legend, Grid and Tight Layout  
--- 

### Title 
--- 

- use `plt.title("title_text")` for adding title to the diagram 


| Parameter    | Meaning              | Example                         |
| ------------ | -------------------- | ------------------------------- |
| `label`      | Title text           | `"Sales Analysis"`              |
| `fontsize`   | Text size            | `fontsize=16`                   | 
| `fontweight` | Font thickness       | `"bold"`                        |
| `fontfamily` | Font family          | `fontfamily="Arial"`            |
| `color`      | Text color           | `color="blue"`                  |
| `loc`        | Position of title    | `"left"`, `"center"`, `"right"` |
| `pad`        | Space above the plot | `pad=15`                        |


```python
# example 
x = [2022, 2023, 2024, 2025, 2026] 
y = [44, 50, 60, 45, 68] 

plt.title(label = "Yearly Data of Students", 
          fontsize = 20, 
          fontweight = "bold", 
          fontfamily = "comic sans ms", 
          color = "#1b3cff", 
          loc = "left", 
          pad = 20)

plt.plot(x, y) 
plt.show()
```

### Axis Labels 
--- 

- use `plt.xlabel("label_text")` and `plt.ylabel("label_text")` for adding lables to the horizontal and vertical axes. 

| Parameter           | Meaning            | Example  |
| ------------------- | ------------------ | -------- |
| `xlabel` / `ylabel` | Label text         | `"Age"`  |
| `fontsize`          | Text size          | `14`     |
| `fontweight`        | Font thickness     | `"bold"` |
| `fontfamily`       | Font family        | `"Arial"`|
| `color`             | Text color         | `"red"`  |
| `labelpad`          | Distance from axis | `10`     |


```python
# example 
x = [2022, 2023, 2024, 2025, 2026] 
y = [44, 50, 60, 45, 68] 

plt.title("Yearly Data Analysis") 

# use dictionary for label styling 
label_style = dict( 
    fontsize = 20, 
    fontweight = "bold", 
    fontfamily = "sans serif",
    color = "#12ff56"  
)

plt.xlabel("Year", **label_style)   # unpack the dictionary 
plt.ylabel(
    "# of Students", 
    fontsize = 24, 
    fontweight = "bold", 
    fontfamily = "comic sans ms", 
    color = "skyblue", 
    labelpad= 20
) 

plt.plot(x, y) 
plt.show()
```

### Legend 
--- 

- A legend explains what each plotted line or dataset represents. 
- use `plt.legend()` for showing the legend. 


- common parameters: 

| Parameter  | Meaning         | Example         |
| ---------- | --------------- | --------------- |
| `loc`      | Legend position | `"upper right"` |
| `fontsize` | Text size       | `12`            |
| `title`    | Legend title    | `"Subjects"`    |
| `frameon`  | Show border     | `True`, `False` |  


- common `loc` values: 

| Value           | Position                       |
| --------------- | ------------------------------ |
| `"best"`        | Automatically chosen (default) |
| `"upper right"` | Top-right                      |
| `"upper left"`  | Top-left                       |
| `"lower right"` | Bottom-right                   |
| `"lower left"`  | Bottom-left                    |
| `"center"`      | Center                         | 



```python
# example 
x1 = [2022, 2023, 2024, 2025, 2026] 
y1 = [44, 50, 60, 45, 68] 
y2 = [55, 48, 56, 60, 45]

plt.title("Yearly Data Analysis") 
plt.xlabel("Year")    
plt.ylabel("# of Students")  


plt.plot(x, y1, label="class A") 
plt.plot(x, y2, label="class B") 

plt.legend()
# plt.legend(
#     title= "classes", 
#     fontsize= 12, 
#     frameon = False, 
#     loc= "upper center")   # further customization  

plt.show()
```

### Grid 
--- 

- A grid adds horizontal and/or vertical reference lines to a plot. 
- making it easier to read and compare values. 
- use `plt.grid()` to add gridlines. 


- common parameters: 

| Parameter           | Meaning                         | Example                |
| ------------------- | ------------------------------- | ---------------------- |
| `axis`              | Which axis to apply the grid to | `'both'`, `'x'`, `'y'` |
| `linestyle` or `ls` | Style of grid lines             | `'--'`, `':'`, `'-'`, `'dashed'`, `'dot'`, `'dashdot'` |
| `linewidth` or `lw` | Thickness of grid lines         | `1`, `2`               |
| `color`             | Grid line color                 | `'gray'`               |
| `alpha`             | Transparency                    | `0.5`                  |
| `visible`           | Show or hide the grid           | `True`, `False`        |


```python
# example 
x = [2022, 2023, 2024, 2025, 2026] 
y = [44, 50, 60, 45, 68] 

plt.title("Yearly Data of Students")
plt.xlabel("year") 
plt.ylabel("# of students")

# plt.grid()
plt.grid( 
    axis= "x", 
    linestyle= "dashed", 
    linewidth= 2, 
    color= "lightgrey", 
    alpha= 0.5
)

plt.plot(x, y) 
plt.show()
```

### Tight Layout 
--- 

- automatically adjusts the spacing between plots so that everything fits neatly.
- use `plt.tight_layout()` before `plt.show()` or `plt.savefig()`. 

```python
# exmaple 
x = np.array([2020, 2021, 2022, 2023, 2024, 2025, 2026])
y1 = np.array([43, 50, 55, 58, 60, 45, 52])
y2 = np.array([48, 50, 51, 48, 54, 60, 49])
y3 = np.array([58, 44, 58, 60, 59, 46, 44])

plt.plot(x, y1, label= "class A")
plt.plot(x, y2, label= "class B")
plt.plot(x, y3, label= "class C")

plt.title("Student Data Analysis", fontsize=32)
plt.xlabel("Years", fontsize= 24) 
plt.ylabel("Number of Students \nin a Class", fontsize= 24) 
plt.legend()

plt.tight_layout()
plt.show()
```

## Marker Customization 
--- 

| Parameter                  | Meaning                           | Example                                                      |
| -------------------------- | --------------------------------- | ------------------------------------------------------------ |
| `marker`                   | Marker shape                      | `'o'`, `'s'`, `'^'`, `'*'`                                   |
| `markersize` or `ms`       | Size of the marker                | `markersize=8`                                               |
| `markerfacecolor` or `mfc` | Fill color of the marker          | `mfc='red'`                                                  |
| `markeredgecolor` or `mec` | Border color of the marker        | `mec='black'`                                                |
| `markeredgewidth` or `mew` | Border thickness                  | `mew=2`                                                      |
| `fillstyle`                | Controls how the marker is filled | `'full'`, `'none'`, `'left'`, `'right'`, `'top'`, `'bottom'` |
| `alpha`                    | Marker transparency               | `alpha=0.6`                                                  |


```python
# use numpy array for faster operations 
x = np.array([2022, 2023, 2024, 2025, 2026])
y1 = np.array([44, 50, 60, 45, 68]) 
y2 = np.array([50, 46, 55, 60, 48]) 

plt.plot(x, y1, marker="s", markerfacecolor="red") 
plt.plot(x, y2, marker="o", markerfacecolor= "green") 
plt.show()
```

### most common marker shapes 
---

| Marker           | Shape           |
| ---------------- | --------------- | 
| `'o'`            | Circle          |
| `'.'`            | Point           |
| `'s'`            | Square          |
| `'^'`            | Triangle Up     |
| `'None'` or `''` | No Marker       |
| `'d'`            | Thin Diamond    |
| `'h'`            | Hexagon         |
| `'v'`            | Triangle Down   |
| `'<'`            | Triangle Left   |
| `'>'`            | Triangle Right  |
| `'D'`            | Diamond         |
| `'p'`            | Pentagon        |
| `'*'`            | Star            |
| `'H'`            | Rotated Hexagon |
| `'+'`            | Plus            |
| `'x'`            | X               |
| `'X'`            | Filled X        |
| `'_'`            | Horizontal Line |


### fillstyle values 
---

| Value      | Effect                  |
| ---------- | ----------------------- |
| `'full'`   | Fully filled (default)  |
| `'none'`   | No fill (hollow marker) |
| `'left'`   | Left half filled        |
| `'right'`  | Right half filled       |
| `'top'`    | Top half filled         |
| `'bottom'` | Bottom half filled      |


```python
# example covering everything 
x1 = np.array([2022, 2023, 2024, 2025, 2026])
y1 = np.array([44, 50, 60, 45, 68]) 

x2 = np.array([2022, 2023, 2024, 2025, 2026])
y2 = np.array([50, 46, 55, 60, 48]) 

# use dictionary for the marker customization 
marker_style = dict( 
    marker = "o", 
    markerfacecolor = "red", 
    markeredgecolor = "black", 
    markersize = 8
)

plt.plot(x1, y1, **marker_style)    # unpack the dictionary  
plt.plot(x2, y2, marker="o", 
                markerfacecolor= "cyan", 
                markeredgecolor = "blue", 
                markersize = 10, 
                markeredgewidth = 2, 
                alpha = 0.5) 
plt.show()
```
