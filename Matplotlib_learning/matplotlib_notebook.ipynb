### Title 
--- 

- use `plt.title("title_text")` for adding title to the diagram 


| Parameter    | Meaning              | Example                         |
| ------------ | -------------------- | ------------------------------- |
| `label`      | Title text           | `"Sales Analysis"`              |
| `fontsize`   | Text size            | `fontsize=16`                   | 
| `fontweight` | Font thickness       | `"bold"`                        |
| `fontfamily` | Font family          | `fontfamily="Arial"`            |
| `color`      | Text color           | `color="blue"`                  |
| `loc`        | Position of title    | `"left"`, `"center"`, `"right"` |
| `pad`        | Space above the plot | `pad=15`                        |


```python
# example 
x = [2022, 2023, 2024, 2025, 2026] 
y = [44, 50, 60, 45, 68] 

plt.title(label = "Yearly Data of Students", 
          fontsize = 20, 
          fontweight = "bold", 
          fontfamily = "comic sans ms", 
          color = "#1b3cff", 
          loc = "left", 
          pad = 20)

plt.plot(x, y) 
plt.show()
```

### Axis Labels 
--- 

- use `plt.xlabel("label_text")` and `plt.ylabel("label_text")` for adding lables to the horizontal and vertical axes. 

| Parameter           | Meaning            | Example  |
| ------------------- | ------------------ | -------- |
| `xlabel` / `ylabel` | Label text         | `"Age"`  |
| `fontsize`          | Text size          | `14`     |
| `fontweight`        | Font thickness     | `"bold"` |
| `fontfamily`       | Font family        | `"Arial"`|
| `color`             | Text color         | `"red"`  |
| `labelpad`          | Distance from axis | `10`     |


```python
# example 
x = [2022, 2023, 2024, 2025, 2026] 
y = [44, 50, 60, 45, 68] 

plt.title("Yearly Data Analysis") 

# use dictionary for label styling 
label_style = dict( 
    fontsize = 20, 
    fontweight = "bold", 
    fontfamily = "sans serif",
    color = "#12ff56"  
)

plt.xlabel("Year", **label_style)   # unpack the dictionary 
plt.ylabel(
    "# of Students", 
    fontsize = 24, 
    fontweight = "bold", 
    fontfamily = "comic sans ms", 
    color = "skyblue", 
    labelpad= 20
) 

plt.plot(x, y) 
plt.show()
```

### Legend 
--- 

- A legend explains what each plotted line or dataset represents. 
- use `plt.legend()` for showing the legend. 


- common parameters: 

| Parameter  | Meaning         | Example         |
| ---------- | --------------- | --------------- |
| `loc`      | Legend position | `"upper right"` |
| `fontsize` | Text size       | `12`            |
| `title`    | Legend title    | `"Subjects"`    |
| `frameon`  | Show border     | `True`, `False` |  


- common `loc` values: 

| Value           | Position                       |
| --------------- | ------------------------------ |
| `"best"`        | Automatically chosen (default) |
| `"upper right"` | Top-right                      |
| `"upper left"`  | Top-left                       |
| `"lower right"` | Bottom-right                   |
| `"lower left"`  | Bottom-left                    |
| `"center"`      | Center                         | 



```python
# example 
x1 = [2022, 2023, 2024, 2025, 2026] 
y1 = [44, 50, 60, 45, 68] 
y2 = [55, 48, 56, 60, 45]

plt.title("Yearly Data Analysis") 
plt.xlabel("Year")    
plt.ylabel("# of Students")  


plt.plot(x, y1, label="class A") 
plt.plot(x, y2, label="class B") 

plt.legend()
# plt.legend(
#     title= "classes", 
#     fontsize= 12, 
#     frameon = False, 
#     loc= "upper center")   # further customization  

plt.show()
```

### Grid 
--- 

- A grid adds horizontal and/or vertical reference lines to a plot. 
- making it easier to read and compare values. 
- use `plt.grid()` to add gridlines. 


- common parameters: 

| Parameter           | Meaning                         | Example                |
| ------------------- | ------------------------------- | ---------------------- |
| `axis`              | Which axis to apply the grid to | `'both'`, `'x'`, `'y'` |
| `linestyle` or `ls` | Style of grid lines             | `'--'`, `':'`, `'-'`, `'dashed'`, `'dot'`, `'dashdot'` |
| `linewidth` or `lw` | Thickness of grid lines         | `1`, `2`               |
| `color`             | Grid line color                 | `'gray'`               |
| `alpha`             | Transparency                    | `0.5`                  |
| `visible`           | Show or hide the grid           | `True`, `False`        |


```python
# example 
x = [2022, 2023, 2024, 2025, 2026] 
y = [44, 50, 60, 45, 68] 

plt.title("Yearly Data of Students")
plt.xlabel("year") 
plt.ylabel("# of students")

# plt.grid()
plt.grid( 
    axis= "x", 
    linestyle= "dashed", 
    linewidth= 2, 
    color= "lightgrey", 
    alpha= 0.5
)

plt.plot(x, y) 
plt.show()
```

### Tight Layout 
--- 

- automatically adjusts the spacing between plots so that everything fits neatly.
- use `plt.tight_layout()` before `plt.show()` or `plt.savefig()`. 

```python
# exmaple 
x = np.array([2020, 2021, 2022, 2023, 2024, 2025, 2026])
y1 = np.array([43, 50, 55, 58, 60, 45, 52])
y2 = np.array([48, 50, 51, 48, 54, 60, 49])
y3 = np.array([58, 44, 58, 60, 59, 46, 44])

plt.plot(x, y1, label= "class A")
plt.plot(x, y2, label= "class B")
plt.plot(x, y3, label= "class C")

plt.title("Student Data Analysis", fontsize=32)
plt.xlabel("Years", fontsize= 24) 
plt.ylabel("Number of Students \nin a Class", fontsize= 24) 
plt.legend()

plt.tight_layout()
plt.show()
```

## Marker Customization 
--- 

- marker highlights the data points ((x, y) values we inserted). 
- use the parameters (given below) inside `ply.plot()` to customize markers. 

### marker customization parameters 
---

| Parameter                  | Meaning                           | Example                                                      |
| -------------------------- | --------------------------------- | ------------------------------------------------------------ |
| `marker`                   | Marker shape                      | `'o'`, `'s'`, `'^'`, `'*'`                                   |
| `markersize` or `ms`       | Size of the marker                | `markersize=8`                                               |
| `markerfacecolor` or `mfc` | Fill color of the marker          | `mfc='red'`                                                  |
| `markeredgecolor` or `mec` | Border color of the marker        | `mec='black'`                                                |
| `markeredgewidth` or `mew` | Border thickness                  | `mew=2`                                                      |
| `fillstyle`                | Controls how the marker is filled | `'full'`, `'none'`, `'left'`, `'right'`, `'top'`, `'bottom'` |
| `alpha`                    | Marker transparency               | `alpha=0.6`                                                  |


```python
# example with mostly used parameters 
# use numpy array for faster operations 
x = np.array([2022, 2023, 2024, 2025, 2026])
y1 = np.array([44, 50, 60, 45, 68]) 
y2 = np.array([50, 46, 55, 60, 48]) 

plt.plot(x, y1, marker="s", 
                markerfacecolor="red", 
                markersize= 8)      # red color square shaped marker 

plt.plot(x, y2, marker="o", 
                markerfacecolor= "green", 
                markersize= 10)     # green color circular shaped marker 

plt.legend(labels= ["class A", "class B"])
plt.show()
```

### most common marker shapes 
---

| Marker           | Shape           |
| ---------------- | --------------- | 
| `'o'`            | Circle          |
| `'.'`            | Point           |
| `'s'`            | Square          |
| `'^'`            | Triangle Up     |
| `'None'` or `''` | No Marker       |
| `'d'`            | Thin Diamond    |
| `'h'`            | Hexagon         |
| `'v'`            | Triangle Down   |
| `'<'`            | Triangle Left   |
| `'>'`            | Triangle Right  |
| `'D'`            | Diamond         |
| `'p'`            | Pentagon        |
| `'*'`            | Star            |
| `'H'`            | Rotated Hexagon |
| `'+'`            | Plus            |
| `'x'`            | X               |
| `'X'`            | Filled X        |
| `'_'`            | Horizontal Line |


### fillstyle values 
---

| Value      | Effect                  |
| ---------- | ----------------------- |
| `'full'`   | Fully filled (default)  |
| `'none'`   | No fill (hollow marker) |
| `'left'`   | Left half filled        |
| `'right'`  | Right half filled       |
| `'top'`    | Top half filled         |
| `'bottom'` | Bottom half filled      |


```python
# example covering everything 
x = np.array([2022, 2023, 2024, 2025, 2026])
y1 = np.array([44, 50, 60, 45, 68]) 
y2 = np.array([50, 46, 55, 60, 48]) 

# use dictionary for marker customization 
marker_style = dict( 
    marker = "o", 
    markerfacecolor = "red", 
    markeredgecolor = "black", 
    markersize = 8
)

plt.plot(x, y1, **marker_style)    # unpack the dictionary  
plt.plot(x, y2, marker="o", 
                markerfacecolor= "cyan", 
                markeredgecolor = "blue", 
                markersize = 10, 
                markeredgewidth = 2, 
                alpha = 0.5) 

plt.show()
```

### Axis Limits 
--- 

- set visible range of the x-axis and y-axis. 

| Function     | Purpose          |

### Tick Locations 
--- 

- ticks = numbers displayed on the axes. 

| Function       | Purpose                   |
| -------------- | ------------------------- |
| `plt.xticks(ticks)` | Set x-axis tick locations |
| `plt.yticks(ticks)` | Set y-axis tick locations |


```python
# example 

x = np.array([2022, 2023, 2024, 2025, 2026])
y = np.array([44, 50, 60, 45, 68]) 

plt.plot(x, y, marker= "o", 
               markersize= 8, 
               markerfacecolor= "skyblue")   

plt.xticks(x) 
plt.yticks(y) 

plt.grid()
plt.show()
```

### Tick Label Customization 
--- 

- customize how the numbers on the axes appear. 
- syntax: `plt.xticks(ticks, labels, fontsize, rotation, color)`

| Parameter  | Meaning            | Example               |
| ---------- | ------------------ | --------------------- |
| `labels`   | Custom tick labels | `["Jan","Feb","Mar"]` |
| `fontsize` | Font size          | `12`                  |
| `fontfamily` | Font family          | `"comic sans ms"`                  |
| `rotation` | Rotate labels      | `45`                  |
| `color`    | Text color         | `"red"`               |


```python
# example 

x = np.array([2022, 2023, 2024, 2025, 2026])
y = np.array([44, 50, 60, 45, 68]) 

plt.plot(x, y, marker= "o", 
               markersize= 8, 
               markerfacecolor= "skyblue")   

# use dictionary for tick label styles
tick_styling = dict(
    labels= ["year 1", "year 2", "year 3", "year 4", "year 5"], 
    fontsize= 12, 
    fontfamily= "comic sans ms", 
    color= "green", 
    rotation= 45
)

plt.xticks(x, **tick_styling)      # unpack the dictionary  
plt.yticks(y, fontsize= 14, color= "red") 

plt.grid()
plt.tight_layout()
plt.show()
```

### Inverting an Axis 
--- 

- syntax: `plt.gca().invert_xaxis()`. 
- `gca()` = get current axis. 

| Function         | Purpose        |
| ---------------- | -------------- |
| `invert_xaxis()` | Reverse x-axis |
| `invert_yaxis()` | Reverse y-axis |


```python
# example 
x = np.array([2022, 2023, 2024, 2025, 2026])
y = np.array([50, 48, 56, 60, 52]) 

plt.xlabel("x axis inverted")
plt.ylabel("y axis inverted")

plt.gca().invert_xaxis()
plt.gca().invert_yaxis()
# plt.gca().set_xinverted(inverted=True)  # can also use this method 

plt.plot(x, y) 
plt.show()
```

## Bar Chart
--- 

- used to compare values across different categories. 
- the height of each bar represents the value of a category. 

- syntax: 
    - `plt.bar(x, height)` for vertical bar chart. 
    - `plt.barh(x, height)` for horizontal bar chart. 
    

most important parameters: 

| Parameter   | Meaning             | Example           |
| ----------- | ------------------- | ----------------- |
| `x`         | Categories (x-axis) | `["A", "B", "C"]` |
| `height`    | Height of each bar  | `[10, 20, 15]`    |
| `width`     | Width of the bars   | `0.5`             |
| `color`     | Fill color          | `"skyblue"`       |
| `edgecolor` | Border color        | `"black"`         |
| `linewidth` | Border thickness    | `2`               |
| `label`     | Legend label        | `"Sales"`         |
| `alpha`     | Transparency        | `0.8`             |


```python
# example 
x = np.array([2022, 2023, 2024, 2025, 2026])
y = np.array([50, 48, 56, 60, 52]) 

plt.xlabel("Year")
plt.ylabel("# of Students")

# use dictionary for bar chart styling 
bar_style = dict(
    color= ["lightgreen", "skyblue", "pink", "grey", "orange"],       
    edgecolor= ["green", "darkblue", "darkred", "black", "chocolate"], 
    width= [0.6, 0.7, 0.8, 0.7, 0.6], 
    label= ["a", "b", "c", "d", "e"]
)   # can also use only one value for all categories (ex: color = "lightgreen" for all bars)

plt.bar(x, y, **bar_style)  # unpack the dictionary  

plt.legend(loc= "upper left")
plt.tight_layout()
plt.show()
```

```python
# example for horizontal bar chart (using plt.barh())
x = np.array([2022, 2023, 2024, 2025, 2026])
y = np.array([50, 48, 56, 60, 52]) 

plt.xlabel("# of Students")
plt.ylabel("Years")
plt.grid(axis= "x", alpha= 0.5)
plt.xticks(y, rotation= 60)

# use dictionary for bar chart styling 
bar_style = dict(
    color= ["lightgreen", "skyblue", "pink", "grey", "orange"],       
    edgecolor= ["green", "darkblue", "darkred", "black", "chocolate"], 
    label= ["a", "b", "c", "d", "e"]
)   # can also use only one value for all categories (ex: color = "lightgreen" for all bars)

plt.barh(x, y, **bar_style)  # unpack the dictionary  

plt.legend(loc= "lower right")
plt.tight_layout()
plt.show()
```

## Pie Chart 
---

- pie chart show how each category contributes to a whole. 
- each slice represents a category. 
- its size is proportional to its value. 
- pie charts are useful for visualizing percentages or proportions. 
- syntax: `plt.pie(x)` 

important parameters: 

| Parameter    | Meaning                           | Example                    |
| ------------ | --------------------------------- | -------------------------- |
| `x`          | Values of each slice              | `[40, 30, 20, 10]`         |
| `labels`     | Labels for slices                 | `["A", "B", "C", "D"]`     |
| `colors`     | Slice colors                      | `["red", "blue", "green"]` |
| `autopct`    | Display percentages               | `"%1.1f%%"`                |
| `startangle` | Starting angle of the first slice | `90`                       |
| `explode`    | Offset one or more slices         | `(0.1, 0, 0, 0)`           |
| `shadow`     | Add shadow                        | `True`                     |


```python
# example 
cost_dict = {
    "Rafi": 230, 
    "Kanon": 110, 
    "Barson": 220, 
    "Sami": 120
}

# plt.pie(cost_dict.values(), labels= cost_dict.keys())   # normal pie chart

# pie chart with customization
plt.pie(
    cost_dict.values(), 
    labels= cost_dict.keys(), 
    autopct="%1.2f%%", 
    explode= [0.05, 0, 0, 0], 
    shadow= True, 
    startangle= 90, 
    colors= ["pink", "lightgreen", "skyblue", "orange"]
)

plt.title("Kabab Party Cost Distribution") 
plt.tight_layout()
plt.show()
```

```python
# another example 
dept_student_dict = {
    "CSE": 181, 
    "EEE": 195, 
    "ME": 181, 
    "CE": 195, 
    "IPE": 120, 
    "ChE": 120, 
    "MME": 120, 
    "BME": 60
} 

plt.pie(
    dept_student_dict.values(), 
    labels= dept_student_dict.keys(), 
    autopct= "%1.2f%%", 
    explode= [0.1, 0, 0, 0, 0, 0, 0, 0], 
    shadow= True
)

plt.title("Students per Department in BUET") 
plt.show()
```

## Scatter Plot 
--- 

- scatter plot displays the relationship between two numerical variables. 
- each point represents one observation. 
- especially valuable for EDA (Exploratory Data Analysis). 
- helps to find relationship, trends, clusters, and outliers in Data Analysis and Machine Learning. 
- syntax: `plt.scatter(x, y)`. 


- mostly used parameters: 

| Parameter | Meaning                        | Example              |
| --------- | ------------------------------ | -------------------- |
| `x`       | X-axis values                  | `[1,2,3,4]`          |
| `y`       | Y-axis values                  | `[5,8,6,9]`          |
| `s`       | Marker size                    | `s=80`               |
| `c`       | Marker color                   | `"red"` or `"green"` |
| `marker`  | Marker style                   | `"o"`, `"^"`, `"s"`  |
| `alpha`   | Transparency                   | `0.7`                |
| `label`   | Legend label                   | `"Data"`             |
| `cmap`    | Colormap (when `c` is numeric) | `"viridis"`          |


```python
# example 
study_hours = np.array([0, 1, 2, 2, 3, 3, 4, 5, 6, 6]) 
marks = np.array([20, 28, 32, 30, 40, 48, 56, 62, 65, 72]) 

# plt.scatter(study_hours, marks, label= "student 1")   # normal scatter plot 

# scatter plot with customization
plt.scatter(
    study_hours, 
    marks, 
    label= "student A", 
    s = 30, 
    c = "green", 
    marker = "s"
) 

plt.title("study hour vs exam marks comparison")
plt.xlabel("hours studied")
plt.ylabel("marks achieved") 

plt.grid(alpha= 0.5)
plt.legend()

plt.tight_layout()
plt.show()
```

### Example with Colormap and Colorbar 
--- 

```python
# example with colormap and colorbar 
study_hours = np.array([0, 1, 2, 2, 3, 3, 4, 5, 6, 6]) 
marks = np.array([20, 28, 32, 30, 40, 48, 56, 62, 65, 72]) 

# plt.scatter(study_hours, marks, label= "student 1")   # normal scatter plot 

# scatter plot with customization
plt.scatter(
    study_hours, 
    marks, 
    label= "student A", 
    marker = "s", 
    s = 50, 
    c = marks, 
    cmap= "viridis", 
) 

plt.colorbar(label= "marks")

plt.title("study hour vs exam marks comparison")
plt.xlabel("hours studied")
plt.ylabel("marks achieved") 

plt.grid(alpha= 0.5)
plt.legend()

plt.tight_layout()
plt.show()
```

## Histograms 
--- 

- histogram show the distribution of a numerical variable by grouping data into intervals. 
- the intervals are called **bins**. 
- the **height** of each bar represents the **number of data points** that fall within a particular interval. 
- difference with bar chart: 
    - **bar chart**: height = amount of that element. 
    - **histograms**: height = number of items in that bin. 
- histogram is used for continuous numerical data, not categories. 
- helps to understand the distribution of features. 

| Parameter   | Meaning                          | Example                 |
| ----------- | -------------------------------- | ----------------------- |
| `x`         | Numerical data                   | `[12, 15, 18, 20, ...]` |
| `bins`      | Number of intervals              | `bins=10`               |
| `color`     | Fill color                       | `"skyblue"`             |
| `edgecolor` | Border color                     | `"black"`               |
| `linewidth` | Border thickness                 | `1.5`                   |
| `density`   | Normalize to probability density | `True`                  |
| `alpha`     | Transparency                     | `0.7`                   |
| `label`     | Legend label                     | `"Age"`                 |


```python
# example 
# generate 100 marks following a normal distribution
marks = np.random.normal(loc= 80, scale= 10, size= 100) # mean is 80, standard deviation is 10 
# restrict the marks to the range [0, 100]
marks = np.clip(marks, a_min= 0, a_max= 100)    

# plt.hist(marks)         # without customization

# with customization
plt.hist(
    marks, 
    bins= 10, 
    color= "lightgreen", 
    edgecolor= "black", 
    linewidth= 2
)

plt.title("Marks Distribution")
plt.xlabel("marks")
plt.ylabel("# of students")

plt.show()
```

```python
x = np.random.randint(low= 0, high= 10000, size= 1000)

plt.hist(x, bins=100) 
plt.show()
```
